### Setting up a Controller class

Let's define a `Controller` class that controls the drone thrusters to reach a given state.  It takes as input part of the state of the drone (`x`,`y`,`theta`), and produces as output the proportion of thrust for the left and right thrusters (a float between 0 and 1).  This specific controller is a Bang Bang controller that pushes up at maximum force if the drone is below its target height, and turns the thrusters off if we are above the target height.

```python
class Controller():
    def __init__(self):
        pass

    def step(self, 
             x, y, theta, 
             dt):        

        if y<0:
            thrust = 1
        else:
            thrust = 0

        thrust_l, thrust_r = thrust, thrust

        return thrust_l, thrust_r # Both values are between 0 and 1
```
